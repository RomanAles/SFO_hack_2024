```python
import os
import pandas as pd
from tqdm import tqdm_notebook as tqdm
```

```python

# Путь к папке с CSV файлами
folder_path = 'C:\\CodesNodes\\HACATONS\\SFO_2024_aut\\Final modeling\\dataset\\data_Q1_2023'

# Столбцы, которые нужно оставить
columns_to_keep = [
    'date', 'serial_number', 'model', 'capacity_bytes', 'failure',
    'smart_5_raw', 'smart_9_raw', 'smart_12_raw', 'smart_187_raw',
    'smart_188_raw', 'smart_190_raw', 'smart_193_raw', 'smart_194_raw',
    'smart_197_raw', 'smart_198_raw', 'smart_199_raw', 'smart_241_raw', 'smart_242_raw'
]

# Инициализация DataFrame для объединения всех данных
combined_data = pd.DataFrame()

# Проходимся по каждому файлу в папке
for file_name in tqdm(os.listdir(folder_path)):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        # Чтение данных из CSV файла с выбранными столбцами
        data = pd.read_csv(file_path, usecols=columns_to_keep)
        # Объединение данных
        combined_data = pd.concat([combined_data, data], ignore_index=True)

print("STEP 1 - OK")

# Сохранение объединенных данных в один CSV файл
combined_data.to_csv('data_Q1_2023.csv', index=False)

print("STEP 2 - OK")

# Фильтрация строк, где failure = 1
failure_data = combined_data[combined_data['failure'] == 1]

# Сохранение данных об отказах в отдельный CSV файл
failure_dates = failure_data[['serial_number', 'date']].rename(columns={'date': 'date_of_failure'})
failure_dates.to_csv('failure_data_Q1_2023.csv', index=False)

print("STEP 3 - OK")
```

```python
data_paths = [
    "data_Q1_2023.csv",
    "data_Q1_2024.csv",
    "data_Q2_2023.csv",
    "data_Q2_2024.csv",
    "data_Q3_2023.csv",
    "data_Q4_2023.csv"
]

df2 = pd.read_csv('failure_data_all.csv')

print("STEP 1 - OK")

for data_path in tqdm(data_paths):

    df1 = pd.read_csv(data_path)

    merged_df = pd.merge(df1, df2, on='serial_number', how='left')

    df = merged_df.dropna(subset=['date_of_failure'])

    df['date'] = pd.to_datetime(df['date'])
    df['date_of_failure'] = pd.to_datetime(df['date_of_failure'])

    df['days_to_failure'] = (df['date_of_failure'] - df['date']).dt.days

    df = df.drop(columns=['date', 'date_of_failure'])

    df.fillna(0, inplace=True)

    print(data_path +" RDY to save")

    df.to_csv("R_" + data_path, index=False)

    print(data_path +" saved")
```
